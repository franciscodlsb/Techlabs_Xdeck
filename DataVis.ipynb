```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
st =  pd.read_csv('combined_startups.csv')
```

```python
# Founding year
sns.distplot(st['Founding Year'], bins = 25)
```

```python
st['Business Model'].value_counts().plot(kind='bar', title='Business Model')
#sns.pairplot(st)
```

```python
for i in st.columns[[4,6,7,8]]:
    plt.figure()
    st[i].value_counts().plot(kind='bar', title = i)
```

```python
#  Most frequent locations and sector
plt.figure()
st['Headquarter'].value_counts()[:10].plot(kind="bar", title ='Headquarter')

plt.figure()
st['Sector'].value_counts()[:10].plot(kind="bar", title = 'Sector')
```
